What is ARIMA?

ARIMA is a model which is used for predicting future trends on a time series data. It is model that form of regression analysis.

AR (Autoregression) : Model that shows a changing variable that regresses on its own lagged/prior values.

I (Integrated) : Differencing of raw observations to allow for the time series to become stationary.

MA (Moving average) : Dependency between an observation and a residual error from a moving average model

For ARIMA models, a standard notation would be ARIMA with p, d, and q, where integer values substitute for the parameters to indicate the type of ARIMA model used.

p: the number of lag observations in the model; also known as the lag order.

d: the number of times that the raw observations are differenced; also known as the degree of differencing.

q: the size of the moving average window; also known as the order of the moving average.

When I took Economotrics, we talked about data being stationary. What does this mean?

When a data is stationary, it simply means that there is no increase or decrease over time. It could be the mean, variance, or standard deviation. For a data to be stationary, the plot of the data should be uniform, looking like an horizontal line.

```python
# Stationary tests
def station(timeseries):
    print('Results of Dickey-Fuller Test:')
    DFtest = adfuller(timeseries, autolag = 'AIC')
    DFoutput = pd.Series(DFtest[0:4], index = ['Test Statistic', 'p-value','#Lags Used', 'Number of Observations Used'])
    for key, value in DFtest[4].items():
        DFoutput['Critical Value (%s)'%key] = value
    print(DFoutput)
    
station(ts)
```

```python
# We need to remove the trend
from pandas import Series as Series
# create a difference series

def differ(data, interval = 1):
    diff = list()
    for i in range(interval, len(data)):
        value = data[i] - data[i - interval]
        diff.append(value)
    return Series(diff)

def inverse_differ(last_ob, value):
    return value + last_ob
```

```python
def tsplot(y, lags = None, 
          figsize = (10, 6), style = 'bmh',
          title = " "):
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):
        fig = plt.figure(figsize = figsize)
        
    layout = (3, 2)
    ts_ax = plt.subplot2grid(layout, (0, 0), 
            colspan = 2)
    acf_ax = plt.subplot2grid(layout, (1, 0))
    pacf_ax = plt.subplot2grid(layout, (1, 1))
    qq_ax = plt.subplot2grid(layout, (2, 0))
    pp_ax = plt.subplot2grid(layout, (2, 1))
    
    y.plot(ax = ts_ax)
    ts_ax.set_title(title)
    smt.graphics.plot_acf(y, lags = lags,
            ax = acf_ax, alpha = 0.5)
    smt.graphics.plot_pacf(y, lags = lags,
            ax = pacf_ax, alpha = 0.5)
    sm.qqplot(y, line = 's', ax = qq_ax)
    qq_ax.set_title('QQ Plot')
    scs.probplot(y, sparams = (y.mean(),
                y.std()), plot = pp_ax)
    plt.tight_layout()

    return
```

```python
np.random.seed(1)
n_samples = int(1000)
a = 0.6
x = w = np.random.normal(size = n_samples)

for t in range(n_samples):
    x[t] = a*x[t-1] + w[t]
limit = 12
_ = tsplot(x, lags = limit, title = "AR(1) process")
```

```python
# Let us do another simulation - AR2 process
n = int(1000)
a = np.array([.444, .333])
b = np.array([0.])
""" Using python Jupyter Notebook, it is required to specify the 
zero-lag value which is 1. Also note that the "a" for the AR
model must be negative. Also is needed to set the "b" for the 
MA = 0 for an AR(p) model"""

ar = np.r_[1, -a]
ma = np.r_[1, b]

ar2 = smt.arma_generate_sample(ar = ar, ma = ma, nsample = n)
_ = tsplot(ar2, lags = 12, title = "AR2 process")
```

```python
# simulate an ARMA(2, 2) model 
# with "a = [0.8, -.65]" and "b = [.5, -.7]"

max_lag = 12

n = int(5000) # taking a big number of samples to estimate

burn = int(n/10) # number of samples to discard before the fit

a = np.array([0.8, -0.65])
b = np.array([0.5, -0.7])

ar = np.r_[1, -a]
ma = np.r_[1, b]

arma22 = smt.arma_generate_sample(ar = ar, ma = ma, nsample = n, burnin = burn)
_ = tsplot(arma22, lags = max_lag, title = "ARMA(2, 2) process")
```

```python
# We are going to use the AIC "Akaike Information Criterion"

best_aic = np.inf
best_order = None
best_model = None

rng = range(5)
for i in rng:
    for j in rng:
        try:
            m_model = smt.ARMA(arma22, order = (i, j)).fit(method = 'mle',
                trend = 'nc')
            m_aic = m_model.aic
            if m_aic < best_aic:
                best_aic = m_aic
                best_order = (i, j)
                best_model = m_model
        except: continue

print('aic: {:6.5f} | order: {}'.format(best_aic, best_order))
```
